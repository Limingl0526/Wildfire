```python
import os
import numpy as np
import pandas as pd

# 设置文件夹路径
folder_path = '/home/limingl/wildfire/CA_model/Output_10/Historical_Land_Use'

# 获取所有csv文件路径，并按顺序排列
file_list = sorted([os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')])

# 读取所有csv文件并存储在矩阵列表中
matrices = [pd.read_csv(file, header=None).values for file in file_list]

# 检查两个矩阵：当前矩阵（current）和下一个矩阵（next）
for i in range(len(matrices) - 1):
    current_matrix = matrices[i]
    next_matrix = matrices[i + 1]

    # 检查第一个矩阵中为0的区域是否在第二个矩阵中为1或2
    zero_to_one_or_two = np.sum((current_matrix == 0) & ((next_matrix == 1) | (next_matrix == 2)))

    # 检查第一个矩阵中为1或2的区域是否在第二个矩阵中为0
    one_or_two_to_zero = np.sum(((current_matrix == 1) | (current_matrix == 2)) & (next_matrix == 0))

    # 打印结果
    print(f"Comparing matrix {i+1} and matrix {i+2}:")
    print(f"Number of elements where current matrix is 0 and next matrix is 1 or 2: {zero_to_one_or_two}")
    print(f"Number of elements where current matrix is 1 or 2 and next matrix is 0: {one_or_two_to_zero}\n")

    # 检查是否 "1 或 2 的区域是否完全包含于另一矩阵中 1 或 2 的位置"
    is_growth_step = np.all(((current_matrix == 1) | (current_matrix == 2)) <= ((next_matrix == 1) | (next_matrix == 2)))

    if is_growth_step:
        print(f"Matrix {i+2} correctly contains all 1 or 2 regions of Matrix {i+1}.\n")
    else:
        print(f"Matrix {i+2} does NOT correctly contain all 1 or 2 regions of Matrix {i+1}.\n")
```
